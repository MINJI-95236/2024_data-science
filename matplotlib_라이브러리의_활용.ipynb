##(1) 그래프에 제목 입력
* title()함수를 통해 그래프에 제목 입력 가능
* 한글깨짐 현상 발생 시 위 코드 복사하여 실행 후 시작

```
plt.title('그래프 제목')
plt.rc('font',family='폰트체입력')
```



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic') #한글 폰트 설정

plt.plot([1,2,3,4,5],[10,30,15,20,40]) #x축, y축 [ ] 리스트로 데이터 각각 입력

plt.title("그래프 제목") #그래프에 제목 넣기
plt.show()
````

## (2) 그래프에 범례 입력
* plot()함수에 label이라는 속성의 레이블 값으로 원하는 문자열을 넣어주고, legend()함수를 실행시키면 레이블 값이 범례로 나타남.


````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic') #한글 폰트 설정

plt.plot([10,30,50,70,90],label='asc') # 해당 그래프 범례의 label 지정
plt.plot([90,70,50,30,10],label='desc') # 해당 그래프 범례의 label 지정

plt.title("범례 입력하기")
plt.legend() # 그래프에 범례 표시하기

plt.show()
````

## (3) 그래프 해상도 설정
* plt.figure()함수 안에서 dpi 속성 또는 figsize 속성을 입력한다.
* dpi=300이란 인치 당 300개의 점이 그래프에 포함된다는 의미를 가진다.


````python
plt.title("그래프 해상도 설정하기(dpi)")

plt.plot([10,30,50,70,90],label='asc') # 해당 그래프 범례의 label 지정, 오름차순
plt.plot([90,70,50,30,10],label='desc') # 해당 그래프 범례의 label 지정, 내림차순
plt.legend() #범례 표시

plt.figure(dpi=300) # 그래프 해상도를 300dpi로 설정

plt.show()
````

````python
plt.title("그래프 해상도 설정하기(figsize)")

plt.plot([10,30,50,70,90],label='asc') # 해당 그래프 범례의 label 지정, 오름차순
plt.plot([90,70,50,30,10],label='desc') # 해당 그래프 범례의 label 지정, 내림차순
plt.legend() #범례 표시

plt.figure(figsize=(8,6)) # 그래프 크기를 8*6인치로 설정

plt.show()
````

## (4) 그래프 축 이름 설정
* plt.xlabel()은 x축 이름을 설정하고, plt.ylabel()은 y축 이름을 설정한다.

````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.plot([1,2,3,4],[1,3,5,7])

plt.title("그래프 축 설정하기")
plt.xlabel('시간') #x축 이름 설정하기
plt.ylabel('학습량') #y축 이름 설정하기

plt.show()
````

## (5) 그래프 색상 지정하기
* 포맷 문자열 사용하기
```
plot(x,y,'r') #red
plot(x,y,'b) #blue
plot(x,y,'g') #green
```
* 색상 키워드 인자 사용하기
```
plot(x,y, color = 'red')
plot(x,y, color = 'blue')
plot(x,y, color = 'green')
```
* 16진수 코드(Hex code) 사용하기
* 색상코드 참고 링크: https://htmlcolorcodes.com/
```
plot(x,y, color = '#FF7F50'
```






````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(4,3))

plt.plot([1,2,3,4],[1,3,5,7], 'r') #포맷 문자열을 사용하여 빨간색으로 지정
plt.plot([1,2,3,4],[0.5,1.0,1.5,2.0],color='blue') #색상 키워드 인자를 사용하여 파란색으로 지정
plt.plot([1,2,3,4],[2,4,6,8],color='#FF7F50') #16진수 코드를 이용하여 살구색으로 지정

plt.title("그래프 색상 지정하기")
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
````

## (6) 그래프 선 스타일 지정하기
* linestyle 속성으로 그래프의 선모양 변경가능
* linestyle 대신 ls 작성가능
```
plt.plot(x,y,ls='--') #선 스타일을 대시 스타일로 지정
plt.plot(x,y,ls=':') #선 스타일을 점 스타일로 지정
```
*marker 속성으로 그래프의 마커모양 변경가능
```
plt.plot(x,y,marker='^') #마커 모양을 삼각형으로 지정
plt.plot(x,y,marker='o') #마커 모양을 동그라미로 지정
plt.plot(x,y,marker='.') #마커 모양을 점으로 지정
```



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(4,3))

plt.plot([1,2,3,4],[1,3,5,7], 'r',ls='--')
plt.plot([1,2,3,4],[0.5,1.0,1.5,2.0],color='blue',ls=':')
plt.plot([1,2,3,4],[2,4,6,8],color='#FF7F50',marker="o",ls='--')

plt.title("그래프 선 스타일 지정하기")
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
````

````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(4,3))
month = ['3월','4월','5월','6월']
grade1=[70,80,65,90]
grade2=[65,82,77,63]


plt.plot(month,grade1,color='r',ls='--',label='국어')
plt.plot(month,grade2,color='b',label='수학')

plt.title("1학기 수행평가 성적")
plt.xlabel('월')
plt.ylabel('점수')
plt.legend()

plt.show()
````

# 3. 다양한 그래프의 활용

## (1) 선그래프



*   시간 흐름에 따른 데이터의 변화를 나타내는데 사용
*   plot() 함수 안에 선그래프로 표현할 x축 데이터, y축 데이터 입력
* x축 데이터 생략하면 y축 데이터에 맞게 0부터 시작되는 정수로 자동 입력



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(8,6))

month = ['3월','4월','5월','6월']
grade=[70,80,65,90]

plt.plot(month,grade,color='r',ls='--',label='국어')

plt.title("1학기 수행평가 성적")
plt.xlabel('월')
plt.ylabel('점수')
plt.legend()

plt.show()
````

##(2) 막대그래프


*   특정 데이터가 나타나는 빈도를 막대 형태로 표현한 것
*   주로 데이터의 크기를 비교하는 목적으로 사용
* 세로 막대그래프: bar()
* 가로 막대그래프: barh()
* plot 함수와 달리 x축과 y축 모두 입력해야 한다.
* width 속성을 활용하여 막대 사이의 너비 설정 가능



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')


plt.figure(figsize=(8,6))

month = ['3월','4월','5월','6월']
grade=[70,80,65,90]

# 컬러를 orange로 설정하고 막대 사이의 너비를 0.5로 설정
plt.bar(month,grade,color='orange',width=0.5,label='국어')
plt.title("1학기 수행평가 성적")
plt.xlabel('월')
plt.ylabel('점수')
plt.legend()

plt.show()
````

## (3) 원그래프(파이차트)


*   주로 특정 항목이 차지하는 비중을 확인하기 위해 사용
* pie() 함수를 사용하여 표현
* labels 속성으로 데이터의 개수만큼 라벨을 나타낼 수 있음.
* autopct 속성으로 부채꼴 안에 표시될 숫자의 형식을 지정 (%.3f%% : 소수 셋째자리까지 표현하라는 의미)
* colors 속성으로 각 영역의 색상 지정 가능(선택 옵션)


````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(8,6))

fruit = ['바나나','배','사과','복숭아']
total_sales=[70,80,65,90]
palette = ['red', 'gold', 'green', 'pink']

plt.pie(total_sales,labels=fruit,autopct='%.2f%%', colors=palette)

plt.title("5월 과일 판매량")

plt.show()
````

## (4) 히스토그램


*   도수 분포표를 그래프로 나타낸 것으로 주로 데이터의 분포를 확인하기 위해 사용
*   hist() 함수로 표현가능
* 입력된 데이터의 빈도를 세로 막대그래프로 표현
* x축은 데이터의 값을 기준으로 자동 생성되므로 입력 시 오류 발생
* bins 속성을 사용하여 x축 구간의 개수 설정 가능(설정하지 않을 경우, 기본 10으로 설정됨)



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(8,6))
weight = [170, 149, 159, 185, 171, 160, 179, 177, 168, 169]

plt.hist(weight, bins = 8) # bins=8는 전체 데이터 범위가 8개의 구간으로 나눠진다.

plt.title("1반 학생들의 키")

plt.show()
````

## (5) 산점도


*   점이 여기저기 흩어진 모양의 그래프로 주로 데이터 분포 확인이나, 데이터 간의 관계 확인을 위해 사용
*   scatter()함수를 통해 표현 가능
* x값과 y값이 만나는 지점을 점으로 찍어서 표현(x축과 y축의 데이터 모두 필요)
* color라는 속성을 이용하여 각각 마커의 색상 지정가능
* alpha라는 속성을 이용하여 마커의 투명도 지정 가능



````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')
plt.figure(figsize=(8,6))

time=[70,80,65,90,73,88,99]
study=[50,60,45,60,53,70,80]

plt.scatter(time, study,color='blue',alpha=0.7) #마커의 색상을 파랑색, 투명도는 70%로 설정

plt.title("1학기 수행평가 성적")
plt.xlabel('시간')
plt.ylabel('학습량')

plt.show()
````

##(6) 상자 그림(상자 수염 그림)


*   주로 데이터의 분포를 확인하기 위해 사용
*   boxplot() 함수를 사용하여 표현 가능
* 데이터를 그대로 표현하는 것이 아니라 데이터에서 통계 자료를 계산하여 나타낸다.
* 최솟값, 하위 25%값, 중앙값, 상위 25%값 최댓값 총 5가지의 값을 표현한다.




````python
import matplotlib.pyplot as plt

plt.rc("font", family='NanumGothic')

plt.figure(figsize=(8,6))

grade1 = [70, 80, 65, 90, 73, 88, 99]
grade2 = [65, 82, 77, 83, 80, 85, 91]

plt.boxplot([grade1, grade2], labels=['국어 점수','수학 점수'])

plt.title("1학기 수행평가 성적")

plt.show()
````
