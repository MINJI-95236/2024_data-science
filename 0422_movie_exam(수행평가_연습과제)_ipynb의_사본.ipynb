# 5. 2019년 이후에 개봉한 영화 데이터 중에서 '영화명'과 '개봉연도' 정보를 출력하세요.

```python
df.loc[df['개봉연도'] >= 2019 ,['영화명','개봉연도']]
```

# 6. 2019년 이후에 개봉한 영화 데이터 중에서 평점이 9.0점 이상인 '영화명'과 '평점' 정보를 출력하세요.

```python
df.loc[(df['개봉연도']>=2019) & (df['평점']>=9.0),['영화명','전국스크린수']]
```
